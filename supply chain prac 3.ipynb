```python
#q3
prob = LpProblem("Simple LP problem",LpMinimize)
x11= LpVariable("x11",0)
x12= LpVariable("x12",0)
x13= LpVariable("x13",0)
x21= LpVariable("x21",0)
x22= LpVariable("x22",0)
x23= LpVariable("x23",0)
x31= LpVariable("x31",0)
x32= LpVariable("x32",0)
x33= LpVariable("x33",0)
```

```python
prob+= x11+x12+x13<=1
prob+= x21+x22+x23<=1
prob+= x31+x32+x33<=1
```

```python
prob+= x11+x21+x31==1
prob+= x12+x22+x32==1
prob+= x13+x23+x33==1
```

```python
prob+= 10*x11+15*x12+9*x13+9*x21+18*x22+5*x23+6*x31+14*x32+3*x33
```

```python
prob.solve()
```

```python
for v in prob.variables():
    print(v.name,"m",v.varValue)
```

```python
print("The optimal value is :",value(prob.objective))
```
